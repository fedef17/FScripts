```python
import os
import sys
import subprocess
```

```python
with open('/nas/archive_CMIP6/lista_wget_deborah.dat', 'rb') as filist:
    linee = [li.strip().split() for li in filist.readlines()]
```

```python
for li in linee:
    # print(str(li[0][1:-1].strip()), str(li[1][1:-1].strip()))
    # print(li[0].decode('utf-8'))
    # print(li[1].decode('utf-8'))

    path = '/nas/archive_CMIP6/' + li[0].decode('utf-8')[1:-1]
    url = li[1].decode('utf-8')[1:-1]

    ind = path.rfind('/')
    cart = path[:ind+1]

    subprocess.run(['mkdir', '-p', cart])
    subprocess.run(['wget', url, '-O', path])
    
```
